```python
#QUEST 04] 파이썬을 사용하여 간단한 계산기 프로그램 만들기
✔️ 문제

#사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만드세요.

✔️ 조건
#1.사용자가 입력한 값이 정수가 아닌 경우
#→ ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다.

#2.정수가 입력될 때까지
#→ "잘못된 입력입니다. 정수를 입력해주세요."를 출력하며
#→ 다시 입력받기를 시도합니다.

#3.나눗셈 연산 시 두 번째 정수가 0인 경우
#→ ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.

#4.사용자가 지원하지 않는 연산자를 입력한 경우
#→ 적절한 오류 메시지를 출력합니다.

#5.math 모듈을 사용하여 제곱 연산(**)을 지원합니다.

#계산기를 통해 계속 계산을 할 것인지 입력받습니다.
```

```python

#import math                # 1. 모듈 불러오고(조건 5)

#def get_integer(prompt):   # 2. 함수 정의(조건 2)


#def calculate(...):        # 3. 함수 정의(연산)

#while True:                # 4. 실제 실행(계산기 실행)
```

```python
import math # 조건 5: math 모듈 임포트

# 조건 2: 정수가 입력될 때까지 반복하는 함수
def get_integer(message):
    while True:
        try:
            value = int(input(message))
            return value
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")#조건 2
```

```python
def calculate(num1, num2, operator):

    if operator == '+':
        return num1 + num2

    elif operator == '-':
        return num1 - num2

    elif operator == '*':
        return num1 * num2

    elif operator == '/':
         # 조건 3: 나눗셈 시 두 번째 정수가 0이면 ZeroDivisionError 처리
        if num2 == 0:
            raise ZeroDivisionError("0으로 나눌 수 없습니다.")
        return num1 / num2

    elif operator == '**':
        # 조건 5: math 모듈을 사용하여 제곱 연산 지원
        return int(math.pow(num1, num2))

    else:
        # 조건 4: 지원하지 않는 연산자 입력 시 오류 메시지 출력
        raise ValueError(f"'{operator}'는 지원하지 않는 연산자입니다.")
```

```python
while True:

    # 두 정수 입력
    num1 = get_integer("첫 번째 정수: ")
    num2 = get_integer("두 번째 정수: ")
    operator = input("연산자 (+,-,*,/,**): ")

    # 계산 실행
    try:
        result = calculate(num1, num2, operator)
        print("결과 =", result)

    except ZeroDivisionError:
        print("오류: 0으로 나눌 수 없습니다.")

    except ValueError:
        print("오류: 지원하지 않는 연산자입니다.")
```
